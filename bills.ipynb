```python
import pandas as pd

# Load product data and drop duplicate IDs to ensure unique index
df_products = pd.read_csv("products.csv")
df_products_unique = df_products.drop_duplicates(subset="ID")

# Create lookup dictionary from unique products
product_lookup = df_products_unique.set_index("ID")[["Product Name", "Brand Name"]].to_dict("index")

# Load billing data (make sure lists are parsed properly)
df_bills = pd.read_csv("bills.csv", converters={"Product_IDs": eval})

# Build 'Products_With_Brands' column
products_with_brands = []

for product_ids in df_bills["Product_IDs"]:
    names = []
    for pid in product_ids:
        info = product_lookup.get(pid, {"Product Name": "Unknown", "Brand Name": "Unknown"})
        names.append(f"{info['Product Name']} ({info['Brand Name']})")
    products_with_brands.append(names)

# Add new column to DataFrame
df_bills["Products_With_Brands"] = products_with_brands

# Save the updated CSV
df_bills.to_csv("bills.csv", index=False)
```
