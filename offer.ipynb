```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import defaultdict
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

class SmartOfferRecommendationSystem:
    def __init__(self):
        # Complete product catalog with all products
        self.products = {
            "Grocery": {
                "01": {"name": "Basmati Rice", "brands": ["Daawat", "India Gate", "Kohinoor", "Tilda", "Fortune"]},
                "02": {"name": "Sona Masoori Rice", "brands": ["Sri Lalitha", "Pride of India", "Fortune", "India Gate", "Tata Sampann"]},
                "03": {"name": "Kolam Rice", "brands": ["India Gate", "Laxmi", "Fortune", "Tata Sampann", "Sri Sri Tattva"]},
                "04": {"name": "Brown Rice", "brands": ["Daawat", "India Gate", "24 Mantra", "Organic Tattva", "Natureland"]},
                "05": {"name": "Matta Rice", "brands": ["Double Horse", "Nirapara", "Eastern", "Bharat Rice", "Pavizham"]},
                "06": {"name": "Wheat Flour", "brands": ["Aashirvaad", "Pillsbury", "Naturefresh", "Patanjali", "ITC"]},
                "07": {"name": "Maida", "brands": ["Pillsbury", "Aashirvaad", "Naturefresh", "Fortune", "Golden Harvest"]},
                "08": {"name": "Multigrain Flour", "brands": ["Aashirvaad", "Patanjali", "Pillsbury", "24 Mantra", "Naturefresh"]},
                "09": {"name": "Pulses", "brands": ["24 Mantra", "Tata Sampann", "PS", "Uttam", "Fortune"]},
                "10": {"name": "Lentils", "brands": ["Tata Sampann", "24 Mantra", "Fortune", "Organic Tattva", "Vedaka"]},
                "11": {"name": "Spices", "brands": ["MDH", "Everest", "Catch", "Badshah", "Patanjali", "Aashirvaad"]},
                "12": {"name": "Masalas", "brands": ["Everest", "MDH", "Catch", "Patanjali", "Aachi", "Badshah"]},
                "13": {"name": "Milk", "brands": ["Amul", "Mother Dairy", "Jersey", "Gopaljee", "Hatsun", "Heritage"]},
                "14": {"name": "Curd", "brands": ["Amul", "Mother Dairy", "Jersey", "Milky Mist", "Gowardhan", "Heritage"]},
                "15": {"name": "Paneer", "brands": ["Amul", "Mother Dairy", "Gowardhan", "Milky Mist", "Nandini", "Heritage"]},
                "16": {"name": "Sunflower Oil", "brands": ["Fortune", "Saffola", "Dhara", "Emami", "Patanjali"]},
                "17": {"name": "Mustard Oil", "brands": ["Patanjali", "Engine", "Fortune", "Dhara", "Emami"]},
                "18": {"name": "Groundnut Oil", "brands": ["Fortune", "24 Mantra", "Dhara", "Natureland", "Anveshan"]},
                "19": {"name": "Soybean Oil", "brands": ["Fortune", "Saffola", "Dhara", "Gagan", "Patanjali"]},
                "20": {"name": "Rice Bran Oil", "brands": ["Fortune", "Saffola", "Riso", "Emami", "Patanjali"]},
                "21": {"name": "Coconut Oil", "brands": ["Parachute", "Maxcare", "KPL Shudhi", "Patanjali", "Vedaka"]},
                "22": {"name": "Olive Oil", "brands": ["Figaro", "Disano", "Borges", "Del Monte", "Oleev"]},
                "23": {"name": "Chips", "brands": ["Lays", "Bingo", "Balaji", "Uncle Chipps", "Too Yumm"]},
                "24": {"name": "Namkeens", "brands": ["Haldiram's", "Bikano", "Bikanervala", "Balaji", "Kurkure"]},
                "25": {"name": "Snacks", "brands": ["Haldiram's", "Parle", "Bikano", "Too Yumm", "Kurkure", "Balaji"]},
                "26": {"name": "Biscuits", "brands": ["Britannia", "Parle", "Sunfeast", "Unibic", "Cremica", "Anmol", "Priya Gold"]},
                "27": {"name": "Chocolates", "brands": ["Amul", "Cadbury", "Nestl√©", "Ferrero", "Campco", "Hershey's India", "Lotus"]},
                "28": {"name": "Tea", "brands": ["Tata Tea", "Brooke Bond", "Girnar", "Society Tea", "Wagh Bakri"]},
                "29": {"name": "Coffee", "brands": ["Bru", "Coffee Day", "Nescaf√©", "Continental", "Tata Coffee"]},
                "30": {"name": "Ready-to-Cook", "brands": ["MTR", "ITC Kitchens of India", "Gits", "Haldiram's", "Kellogg's"]},
                "31": {"name": "Ready-to-Eat", "brands": ["MTR", "ITC Kitchens of India", "Gits", "Haldiram's", "Kellogg's"]},
                "32": {"name": "Frozen Paratha", "brands": ["ITC Master Chef", "Amul", "McCain", "Haldiram's", "Vadilal"]},
                "33": {"name": "Frozen French Fries", "brands": ["McCain", "ITC Master Chef", "Godrej Yummiez", "Vadilal"]},
                "34": {"name": "Frozen Vegetables Mix", "brands": ["Safal", "Vadilal", "Godrej Yummiez", "Mother Dairy"]},
                "35": {"name": "Frozen Samosa", "brands": ["Haldiram's", "ITC Master Chef", "McCain", "Vadilal"]},
                "36": {"name": "Frozen Paneer Tikka", "brands": ["ITC Master Chef", "Godrej Yummiez", "Vadilal", "Amul"]},
                "37": {"name": "Frozen Nuggets", "brands": ["Godrej Yummiez", "ITC Master Chef", "McCain", "Venky's"]},
                "38": {"name": "Noodles", "brands": ["Maggi", "Sunfeast Yippee", "Top Ramen", "Wai Wai", "Knorr", "Patanjali"]},
                "39": {"name": "Semiya (Vermicelli)", "brands": ["MTR", "Bambino", "Aachi", "Anil", "Double Horse"]},
                "40": {"name": "Pasta", "brands": ["Weikfield", "Disano", "Bambino", "Del Monte", "Sunfeast", "Barilla"]},
                "41": {"name": "Spreads", "brands": ["Nutella", "Pintola", "Dr. Oetker", "Amul", "Alpino", "Veeba"]},
                "42": {"name": "Sauces", "brands": ["Kissan", "Del Monte", "Veeba", "Dr. Oetker", "FunFoods", "Maggi"]}
            },
            "Clothing": {
                "01": {"name": "T-Shirt", "brands": ["Peter England", "Allen Solly", "Louis Philippe", "Roadster", "UCB", "John Players"]},
                "02": {"name": "Shirts", "brands": ["Louis Philippe", "Peter England", "Raymond", "Van Heusen", "Allen Solly", "UCB"]},
                "03": {"name": "Jeans", "brands": ["Levi's", "Wrangler", "Spykar", "Lee", "Flying Machine", "Pepe Jeans"]},
                "04": {"name": "Dresses", "brands": ["Biba", "W for Woman", "Global Desi", "AND", "FabIndia", "Zara"]},
                "05": {"name": "Salwar Suits", "brands": ["Biba", "W for Woman", "Aurelia", "Global Desi", "FabIndia", "Libas"]},
                "06": {"name": "Ethnic Wear", "brands": ["Manyavar", "Mohey", "FabIndia", "Biba", "W for Woman", "Global Desi"]},
                "07": {"name": "Kurtas", "brands": ["FabIndia", "Manyavar", "Aurelia", "Libas", "Indya", "Soch"]},
                "08": {"name": "Leggings", "brands": ["Lyra", "Jockey", "Rupa", "Zivame", "Biba", "Aurelia"]},
                "09": {"name": "Track Pants", "brands": ["Jockey", "Puma", "Nike", "HRX", "Adidas", "Sparx"]},
                "10": {"name": "Joggers", "brands": ["Roadster", "HRX", "Bewakoof", "Adidas", "Nike", "Levi's"]},
                "11": {"name": "Shorts", "brands": ["HRX", "Nike", "Jockey", "Bewakoof", "Roadster", "UCB"]},
                "12": {"name": "Hoodies", "brands": ["Puma", "Adidas", "HRX", "Roadster", "Nike", "H&M"]},
                "13": {"name": "Blazers", "brands": ["Van Heusen", "Peter England", "Raymond", "Louis Philippe", "Allen Solly", "Zara"]},
                "14": {"name": "Formal Trousers", "brands": ["Raymond", "Van Heusen", "Peter England", "Louis Philippe", "Blackberrys", "Allen Solly"]},
                "15": {"name": "Sarees", "brands": ["Suta", "FabIndia", "Bunkar", "Kalpana", "Nalli", "Varkala Silk"]},
                "16": {"name": "Lehengas", "brands": ["Manyavar", "Mohey", "Biba", "W for Woman", "Koskii", "Libas"]},
                "17": {"name": "Jackets", "brands": ["Puma", "Nike", "Roadster", "H&M", "Zara", "UCB"]},
                "18": {"name": "Skirts", "brands": ["Zara", "AND", "FabIndia", "UCB", "Global Desi", "Forever 21"]},
                "19": {"name": "Jumpsuits", "brands": ["AND", "Global Desi", "Only", "Forever 21", "Mango", "Zara"]},
                "20": {"name": "Nightwear", "brands": ["Jockey", "Clovia", "Zivame", "Van Heusen", "Bummer", "Enamor"]}
            },
            "Sports & Toys": {
                "01": {"name": "Cricket Bat", "brands": ["SG", "SS", "MRF", "Kookaburra", "CEAT", "New Balance"]},
                "02": {"name": "Football", "brands": ["Nivia", "Adidas", "Nike", "Puma", "Select", "Vector X"]},
                "03": {"name": "Board Game", "brands": ["Funskool", "Hasbro", "Mattel", "WinGo", "Frank", "Smartivity"]},
                "04": {"name": "Cycle", "brands": ["Hero", "Atlas", "Firefox", "Montra", "Hercules", "Ninety One"]},
                "05": {"name": "Yoga Mat", "brands": ["Decathlon", "HRX", "Reebok", "Nike", "Cultsport", "Boldfit"]},
                "06": {"name": "Swim Suits", "brands": ["Decathlon", "Speedo", "Arena", "HRX", "Nivia", "Zivame"]},
                "07": {"name": "Skating Shoes", "brands": ["Nivia", "Decathlon", "Strauss", "KRF", "Jaspo", "Cosco"]},
                "08": {"name": "Hot Wheels", "brands": ["Hot Wheels", "Maisto", "Mattel", "Majorette", "Toyzone", "Speedage"]},
                "09": {"name": "Nerf Guns", "brands": ["Nerf", "BoomCo", "X-Shot", "Adventure Force", "Buzz Bee", "Hasbro"]},
                "10": {"name": "Badminton Racket", "brands": ["Yonex", "Li-Ning", "Cosco", "Nivia", "Silver's", "Victor"]},
                "11": {"name": "Cricket Ball", "brands": ["SG", "SS", "Kookaburra", "DSC", "GM", "BDM"]},
                "12": {"name": "Basketball", "brands": ["Cosco", "Nivia", "Spalding", "Adidas", "Nike", "Puma"]},
                "13": {"name": "Tennis Racket", "brands": ["Wilson", "Yonex", "Head", "Babolat", "Tecnifibre", "Decathlon"]},
                "14": {"name": "Dart Board", "brands": ["Strauss", "Aurion", "Winmau", "Toyshine", "Shot", "Harrows"]},
                "15": {"name": "Skipping Rope", "brands": ["Boldfit", "Strauss", "Nivia", "Cosco", "HRX", "Decathlon"]},
                "16": {"name": "Toy Car Remote", "brands": ["Toyshine", "RemoteX", "Maisto", "Speedage", "R/C Zone", "Hikoki"]},
                "17": {"name": "Action Figures", "brands": ["Hasbro", "Marvel", "Funskool", "Toyshine", "DC", "Transformers"]},
                "18": {"name": "Doll Set", "brands": ["Barbie", "Funskool", "Toyshine", "Mattel", "Chicco", "Zephyr"]},
                "19": {"name": "Tricycle", "brands": ["R for Rabbit", "LuvLap", "Amardeep", "Baybee", "Dash", "Toyhouse"]},
                "20": {"name": "Scooter (Kick)", "brands": ["R for Rabbit", "Strauss", "Toyshine", "Aurion", "NHR", "Zyklo"]}
            },
            "Electronics": {
                "01": {"name": "Headphones", "brands": ["boAt", "JBL", "Sony", "Noise", "Sennheiser", "Realme"]},
                "02": {"name": "LED TV", "brands": ["Samsung", "Sony", "LG", "Panasonic", "TCL", "Mi"]},
                "03": {"name": "Mixer Grinder", "brands": ["Philips", "Preethi", "Bajaj", "Sujata", "Panasonic", "Havells"]},
                "04": {"name": "Refrigerator", "brands": ["Samsung", "LG", "Whirlpool", "Haier", "Godrej", "Panasonic"]},
                "05": {"name": "Microwave", "brands": ["Samsung", "IFB", "Godrej", "LG", "Panasonic", "Haier"]},
                "06": {"name": "Beard Trimmer", "brands": ["Philips", "Syska", "Mi", "Havells", "Nova", "Wahl"]},
                "07": {"name": "Smartphone", "brands": ["Samsung", "OnePlus", "Xiaomi", "Realme", "Vivo", "Apple"]},
                "08": {"name": "Tablet", "brands": ["Samsung", "Apple", "Lenovo", "Xiaomi", "Realme", "Honor"]},
                "09": {"name": "Smartwatch", "brands": ["Noise", "boAt", "Fire-Boltt", "Realme", "Samsung", "Apple"]},
                "10": {"name": "Bluetooth Speaker", "brands": ["JBL", "boAt", "Sony", "Philips", "Marshall", "Zebronics"]},
                "11": {"name": "Laptop", "brands": ["HP", "Dell", "Lenovo", "Asus", "Acer", "Apple"]},
                "12": {"name": "Printer", "brands": ["HP", "Canon", "Epson", "Brother", "Samsung", "Ricoh"]},
                "13": {"name": "Monitor", "brands": ["Samsung", "LG", "Dell", "HP", "Acer", "Asus"]},
                "14": {"name": "Camera", "brands": ["Canon", "Nikon", "Sony", "Fujifilm", "GoPro", "Panasonic"]},
                "15": {"name": "Power Bank", "brands": ["Mi", "Realme", "Ambrane", "Samsung", "URBN", "Syska"]},
                "16": {"name": "WiFi Router", "brands": ["TP-Link", "Netgear", "D-Link", "Tenda", "ASUS", "Xiaomi"]},
                "17": {"name": "External Hard Drive", "brands": ["WD", "Seagate", "Toshiba", "Sony", "Lenovo", "Transcend"]},
                "18": {"name": "Pendrive", "brands": ["SanDisk", "HP", "Sony", "Strontium", "Kingston", "Transcend"]},
                "19": {"name": "VR Headset", "brands": ["Oculus", "HTC", "Sony", "Irusu", "Procus", "DPVR"]},
                "20": {"name": "Gaming Console", "brands": ["Sony PlayStation", "Xbox", "Nintendo", "Zebronics", "Croma", "Lenovo Legion"]}
            },
            "Home Appliances": {
                "01": {"name": "Refrigerator", "brands": ["Godrej", "Whirlpool", "LG", "Samsung", "Haier", "Panasonic"]},
                "02": {"name": "Microwave", "brands": ["IFB", "Samsung", "LG", "Godrej", "Panasonic", "Haier"]},
                "03": {"name": "Mixer Grinder", "brands": ["Philips", "Preethi", "Bajaj", "Panasonic", "Sujata", "Havells"]},
                "04": {"name": "Washing Machine", "brands": ["LG", "Samsung", "Whirlpool", "IFB", "Bosch", "Haier"]},
                "05": {"name": "Water Purifier", "brands": ["Kent", "Aquaguard", "Pureit", "Livpure", "Blue Star", "AO Smith"]},
                "06": {"name": "Air Conditioner", "brands": ["Voltas", "LG", "Samsung", "Blue Star", "Daikin", "Hitachi"]},
                "07": {"name": "Ceiling Fan", "brands": ["Havells", "Orient", "Usha", "Crompton", "Bajaj", "Anchor"]},
                "08": {"name": "Room Heater", "brands": ["Bajaj", "Usha", "Orpat", "Morphy Richards", "Havells", "V-Guard"]},
                "09": {"name": "Electric Kettle", "brands": ["Prestige", "Philips", "Pigeon", "Havells", "Bajaj", "Kent"]},
                "10": {"name": "Induction Cooktop", "brands": ["Prestige", "Philips", "Pigeon", "Bajaj", "Usha", "Butterfly"]},
                "11": {"name": "Dishwasher", "brands": ["Bosch", "IFB", "Voltas Beko", "LG", "Samsung", "Siemens"]},
                "12": {"name": "Kitchen Chimney", "brands": ["Elica", "Faber", "Glen", "Hindware", "Kaff", "Sunflame"]},
                "13": {"name": "Air Purifier", "brands": ["Philips", "Dyson", "Sharp", "Honeywell", "Mi", "Blue Star"]},
                "14": {"name": "Geyser", "brands": ["AO Smith", "Racold", "Havells", "Bajaj", "V-Guard", "Crompton"]},
                "15": {"name": "OTG Oven", "brands": ["Morphy Richards", "Bajaj", "Prestige", "Wonderchef", "Philips", "Usha"]},
                "16": {"name": "Vacuum Cleaner", "brands": ["Eureka Forbes", "Dyson", "Philips", "Panasonic", "Agaro", "Kent"]},
                "17": {"name": "Toaster", "brands": ["Philips", "Bajaj", "Prestige", "Morphy Richards", "Havells", "Pigeon"]},
                "18": {"name": "Juicer", "brands": ["Philips", "Sujata", "Bajaj", "Havells", "Morphy Richards", "Panasonic"]},
                "19": {"name": "Steam Iron", "brands": ["Philips", "Bajaj", "Havells", "Morphy Richards", "Usha", "Crompton"]},
                "20": {"name": "Electric Rice Cooker", "brands": ["Prestige", "Panasonic", "Bajaj", "Pigeon", "Kent", "Butterfly"]}
            },
            "Cosmetics": {
                "01": {"name": "Lipstick", "brands": ["Lakm√©", "Maybelline", "L'Or√©al", "Colorbar", "Revlon", "Faces Canada"]},
                "02": {"name": "Primer", "brands": ["Maybelline New York", "L'Or√©al Paris", "Lakm√©", "Colorbar", "Revlon", "Sugar Cosmetics"]},
                "03": {"name": "Concealer", "brands": ["Lakm√©", "Colorbar", "Sugar Cosmetics", "Insight Cosmetics", "Blue Heaven", "Mars Cosmetics"]},
                "04": {"name": "Foundation", "brands": ["Colorbar", "Sugar Cosmetics", "Insight Cosmetics", "Blue Heaven", "Mars Cosmetics", "MyGlamm"]},
                "05": {"name": "Eyeliner", "brands": ["Lakm√©", "Colorbar", "Sugar Cosmetics", "Blue Heaven", "Insight Cosmetics", "Mars Cosmetics"]},
                "06": {"name": "Compact", "brands": ["Lakm√©", "Colorbar", "Sugar Cosmetics", "Blue Heaven", "Insight Cosmetics", "Mars Cosmetics"]},
                "07": {"name": "Blush", "brands": ["Lakm√©", "MyGlamm", "Insight Cosmetics", "Mars Cosmetics", "Blue Heaven", "Swiss Beauty"]},
                "08": {"name": "Kajal", "brands": ["Lakm√©", "Himalaya", "Faces Canada", "Sugar Cosmetics", "Blue Heaven", "Elle 18"]},
                "09": {"name": "Makeup Remover", "brands": ["Garnier", "Lakm√©", "Nivea", "L'Or√©al Paris", "Simple", "Bioderma"]},
                "10": {"name": "BB Cream", "brands": ["Pond's", "Garnier", "Lakm√©", "Fair & Lovely", "Maybelline", "Colorbar"]}
            },
            "Home & Kitchen": {
                "01": {"name": "Cookware", "brands": ["Prestige", "Pigeon", "Wonderchef", "Hawkins", "Butterfly", "Vinod"]},
                "02": {"name": "Gas Stove", "brands": ["Prestige", "Sunflame", "Butterfly", "Elica", "Glen", "Pigeon"]},
                "03": {"name": "Pressure Cooker", "brands": ["Hawkins", "Prestige", "Butterfly", "Pigeon", "Borosil", "Vinod"]},
                "04": {"name": "Water Bottle", "brands": ["Milton", "Cello", "Tupperware", "Borosil", "Signoraware", "Pigeon"]},
                "05": {"name": "Dinner Set", "brands": ["La Opala", "Borosil", "Corelle", "Clay Craft", "Cello", "Neelam"]},
                "06": {"name": "Kitchen Storage", "brands": ["Tupperware", "Milton", "Signoraware", "Pigeon", "Cello", "Borosil"]},
                "07": {"name": "Dustbin", "brands": ["Nilkamal", "Cello", "Spotzero", "Bathla", "Prestige", "Kuber Industries"]},
                "08": {"name": "Chopper", "brands": ["Pigeon", "Prestige", "Wonderchef", "Butterfly", "Borosil", "Inalsa"]},
                "09": {"name": "Knife Set", "brands": ["Victorinox", "Prestige", "Pigeon", "Wonderchef", "Butterfly", "Borosil"]},
                "10": {"name": "Serving Bowl", "brands": ["La Opala", "Borosil", "Corelle", "Cello", "Clay Craft", "Neelam"]}
            },
            "Footwear": {
                "01": {"name": "Men's Sports Shoes", "brands": ["Nike", "Adidas", "Puma", "Reebok", "Campus", "Sparx"]},
                "02": {"name": "Men's Casual Shoes", "brands": ["Bata", "Red Tape", "Woodland", "Lee Cooper", "Sparx", "Hush Puppies"]},
                "03": {"name": "Men's Sandals", "brands": ["Relaxo", "Sparx", "Bata", "Paragon", "Flite", "Red Chief"]},
                "04": {"name": "Men's Formal Shoes", "brands": ["Hush Puppies", "Bata", "Red Tape", "Lee Cooper", "Clarks", "Woodland"]},
                "05": {"name": "Men's Flip Flops", "brands": ["Sparx", "Puma", "Adidas", "Nike", "Paragon", "Relaxo"]},
                "06": {"name": "Women's Heels", "brands": ["Catwalk", "Metro", "Mochi", "Inc.5", "Bata", "Clarks"]},
                "07": {"name": "Women's Flats", "brands": ["Bata", "Catwalk", "Metro", "Paragon", "Mochi", "Relaxo"]},
                "08": {"name": "Women's Sports Shoes", "brands": ["Nike", "Adidas", "Puma", "Reebok", "Campus", "Skechers"]},
                "09": {"name": "Women's Casual Shoes", "brands": ["Bata", "Skechers", "Red Tape", "Metro", "Woodland", "Crocs"]},
                "10": {"name": "Children's Footwear", "brands": ["Bata", "Disney", "Crocs", "Sparx", "Paragon", "Campus"]}
            },
            "Bath & Body": {
                "01": {"name": "Body Lotion", "brands": ["Nivea", "Vaseline", "Parachute", "Dove", "Himalaya", "Mamaearth"]},
                "02": {"name": "Shower Gel", "brands": ["Nivea", "Dove", "Palmolive", "Fiama", "The Body Shop", "Himalaya"]},
                "03": {"name": "Soap", "brands": ["Lux", "Dove", "Pears", "Dettol", "Lifebuoy", "Cinthol"]},
                "04": {"name": "Shampoo", "brands": ["Pantene", "Head & Shoulders", "Dove", "Tresemme", "Sunsilk", "Clinic Plus"]},
                "05": {"name": "Hair Oil", "brands": ["Parachute", "Dabur", "Indulekha", "Patanjali", "Bajaj", "Emami"]},
                "06": {"name": "Face Wash", "brands": ["Himalaya", "Clean & Clear", "Garnier", "Pond's", "Mamaearth", "Nivea"]},
                "07": {"name": "Handwash", "brands": ["Dettol", "Lifebuoy", "Savlon", "Himalaya", "Palmolive", "Godrej"]},
                "08": {"name": "Sanitizer", "brands": ["Dettol", "Lifebuoy", "Savlon", "Himalaya", "Sterillium", "Purell"]},
                "09": {"name": "Toothpaste", "brands": ["Colgate", "Pepsodent", "Close-Up", "Sensodyne", "Patanjali", "Dabur"]},
                "10": {"name": "Toothbrush", "brands": ["Colgate", "Oral-B", "Close-Up", "Sensodyne", "Trisa", "Ajanta"]}
            }
        }

        # Realistic price ranges for each product
        self.price_ranges = {
            "Grocery": {"01": (50, 200), "02": (50, 200), "03": (30, 150), "04": (40, 180), "05": (20, 100), "06": (10, 150), "07": (30, 300), "08": (20, 200), "09": (40, 120), "10": (40, 120), "11": (10, 100), "12": (10, 100), "13": (20, 80), "14": (20, 60), "15": (50, 200), "16": (100, 300), "17": (100, 300), "18": (120, 350), "19": (100, 300), "20": (120, 350), "21": (50, 200), "22": (200, 600), "23": (10, 60), "24": (10, 60), "25": (10, 100), "26": (5, 100), "27": (10, 300), "28": (20, 200), "29": (50, 300), "30": (50, 300), "31": (50, 300), "32": (80, 250), "33": (60, 200), "34": (50, 200), "35": (80, 250), "36": (100, 300), "37": (100, 300), "38": (10, 100), "39": (10, 60), "40": (20, 150), "41": (100, 400), "42": (30, 150)},
            "Clothing": {"01": (300, 1200), "02": (500, 2000), "03": (1000, 3500), "04": (1000, 4000), "05": (1000, 3500), "06": (1500, 6000), "07": (500, 2500), "08": (200, 700), "09": (500, 2000), "10": (600, 2500), "11": (300, 1500), "12": (900, 3500), "13": (2000, 7000), "14": (800, 3000), "15": (1000, 10000), "16": (1500, 12000), "17": (1200, 5000), "18": (700, 3000), "19": (1200, 4000), "20": (500, 2000)},
            "Sports & Toys": {"01": (1000, 8000), "02": (500, 3000), "03": (200, 2000), "04": (5000, 20000), "05": (300, 2000), "06": (300, 2500), "07": (1000, 5000), "08": (200, 3000), "09": (500, 4000), "10": (500, 4000), "11": (300, 1500), "12": (500, 3000), "13": (1000, 10000), "14": (300, 2500), "15": (150, 1000), "16": (400, 3000), "17": (300, 2500), "18": (300, 2000), "19": (1000, 7000), "20": (1000, 4000)},
            "Electronics": {"01": (500, 15000), "02": (10000, 200000), "03": (2000, 10000), "04": (15000, 100000), "05": (5000, 20000), "06": (500, 5000), "07": (5000, 90000), "08": (7000, 80000), "09": (1000, 30000), "10": (1000, 25000), "11": (20000, 200000), "12": (3000, 40000), "13": (5000, 100000), "14": (20000, 300000), "15": (500, 4000), "16": (1000, 10000), "17": (3000, 15000), "18": (300, 3000), "19": (2500, 80000), "20": (20000, 80000)},
            "Home Appliances": {"01": (15000, 100000), "02": (5000, 20000), "03": (2000, 10000), "04": (10000, 50000), "05": (5000, 25000), "06": (25000, 60000), "07": (1200, 5000), "08": (1000, 7000), "09": (600, 3000), "10": (1500, 6000), "11": (20000, 60000), "12": (5000, 25000), "13": (5000, 45000), "14": (2000, 15000), "15": (3000, 12000), "16": (3000, 60000), "17": (800, 3500), "18": (1500, 8000), "19": (800, 3500), "20": (1500, 6000)},
            "Cosmetics": {"01": (200, 1200), "02": (250, 1000), "03": (200, 900), "04": (300, 1500), "05": (150, 700), "06": (150, 600), "07": (200, 800), "08": (100, 400), "09": (150, 800), "10": (150, 700)},
            "Home & Kitchen": {"01": (500, 4000), "02": (1500, 8000), "03": (800, 3500), "04": (200, 1500), "05": (1000, 7000), "06": (300, 2500), "07": (300, 3000), "08": (300, 2000), "09": (500, 3500), "10": (300, 2500)},
            "Footwear": {"01": (1000, 7000), "02": (800, 5000), "03": (400, 2500), "04": (1000, 6000), "05": (200, 2000), "06": (700, 4000), "07": (300, 2500), "08": (1000, 7000), "09": (800, 4000), "10": (300, 2500)},
            "Bath & Body": {"01": (100, 500), "02": (100, 400), "03": (20, 100), "04": (100, 700), "05": (50, 300), "06": (100, 400), "07": (50, 300), "08": (20, 150), "09": (20, 150), "10": (20, 100)}
        }

        # Customer purchase history for bill analysis
        self.customer_bills = []

        # Initialize product data with sales metrics
        self._initialize_product_data()

    def _initialize_product_data(self):
        """Initialize product data with sales metrics and offer potential"""
        self.product_data = {}

        for category, products in self.products.items():
            self.product_data[category] = {}
            for product_id, product_info in products.items():
                price_min, price_max = self.price_ranges[category][product_id]

                brand_data = {}
                for brand in product_info['brands']:
                    # Generate realistic sales data
                    units_sold = random.randint(100, 2000)
                    base_price = random.randint(price_min, price_max)
                    rating = round(random.uniform(3.0, 4.9), 1)
                    reviews = random.randint(50, 1000)

                    # Calculate offer metrics
                    discount_potential = random.randint(5, 40)  # 5-40% discount potential
                    seasonal_boost = random.uniform(0.8, 1.5)
                    brand_popularity = random.uniform(0.6, 1.4)

                    brand_data[brand] = {
                        'price': base_price,
                        'units_sold': units_sold,
                        'revenue': units_sold * base_price,
                        'rating': rating,
                        'reviews': reviews,
                        'discount_potential': discount_potential,
                        'seasonal_boost': seasonal_boost,
                        'brand_popularity': brand_popularity,
                        'stock_level': random.randint(20, 500)
                    }

                self.product_data[category][product_id] = {
                    'name': product_info['name'],
                    'brands': brand_data
                }

    def display_main_menu(self):
        """Display enhanced main menu with offer options"""
        print("\n" + "="*80)
        print("        üéØ SMART E-COMMERCE OFFER RECOMMENDATION SYSTEM üéØ")
        print("="*80)
        print("Choose your shopping experience:")
        print("-" * 50)

        categories = list(self.products.keys())
        for i, category in enumerate(categories, 1):
            total_products = len(self.products[category])
            print(f"{i}. {category:<20} ({total_products} products)")

        print("-" * 50)
        print("9. üìä Complete Analytics Dashboard")
        print("8. üßæ Bill Analysis & Smart Recommendations")
        print("7. üéÅ Today's Special Offers")
        print("6. üîç Smart Product Search")
        print("0. Exit")
        print("="*80)

    def display_category_products(self, category):
        """Display products with offer indicators"""
        print(f"\nüõçÔ∏è {category.upper()} COLLECTION - BEST OFFERS INSIDE!")
        print("="*70)

        products = self.products[category]
        for product_id, product_info in products.items():
            brands_count = len(product_info['brands'])
            price_range = self.price_ranges[category][product_id]

            # Calculate best offer for this product
            best_offer = self._calculate_best_offer(category, product_id)
            offer_indicator = f"üî• {best_offer}% OFF" if best_offer > 15 else f"üí∞ {best_offer}% OFF"

            print(f"{product_id}. {product_info['name']:<30} ({brands_count} brands) "
                  f"‚Çπ{price_range[0]}-{price_range[1]} {offer_indicator}")

        print("="*70)
        print("0. ‚¨ÖÔ∏è Back to main menu")

    def _calculate_best_offer(self, category, product_id):
        """Calculate the best available offer for a product"""
        brand_data = self.product_data[category][product_id]['brands']
        best_discount = max(data['discount_potential'] for data in brand_data.values())
        return int(best_discount)

    def display_product_offers(self, category, product_id):
        """Display comprehensive product analysis with attractive offers"""
        product = self.product_data[category][product_id]
        product_name = product['name']

        print(f"\nüéØ {product_name.upper()} - EXCLUSIVE OFFERS & ANALYSIS")
        print("="*80)

        # Prepare brand data for analysis
        brand_offers = []
        for brand, data in product['brands'].items():
            original_price = data['price']
            discount = data['discount_potential']
            offer_price = original_price * (1 - discount/100)
            savings = original_price - offer_price

            brand_offers.append({
                'Brand': brand,
                'Original_Price': original_price,
                'Offer_Price': round(offer_price),
                'Discount': discount,
                'Savings': round(savings),
                'Rating': data['rating'],
                'Reviews': data['reviews'],
                'Stock': data['stock_level'],
                'Units_Sold': data['units_sold']
            })

        # Sort by best offers (highest discount)
        brand_offers.sort(key=lambda x: x['Discount'], reverse=True)

        # Display offer table
        print("üè∑Ô∏è BRAND OFFERS COMPARISON:")
        print("-" * 80)
        print(f"{'Brand':<15} {'Original':<10} {'Offer':<10} {'Discount':<10} {'Savings':<10} {'Rating':<8}")
        print("-" * 80)

        for offer in brand_offers:
            print(f"{offer['Brand']:<15} ‚Çπ{offer['Original_Price']:<9} ‚Çπ{offer['Offer_Price']:<9} "
                  f"{offer['Discount']}%{'':<7} ‚Çπ{offer['Savings']:<9} {offer['Rating']}/5")

        # Create attractive visualizations
        self._create_offer_analysis_charts(product_name, brand_offers)

        # Provide smart recommendations
        self._generate_smart_recommendations(product_name, brand_offers)

        # Add to customer consideration (simulated shopping cart)
        self._add_to_consideration(category, product_id, brand_offers)

    def _create_offer_analysis_charts(self, product_name, brand_offers):
        """Create beautiful and informative charts for offer analysis"""
        plt.style.use('dark_background')
        fig = plt.figure(figsize=(18, 12))
        fig.suptitle(f'{product_name} - Smart Offer Analysis Dashboard',
                     fontsize=20, fontweight='bold', color='white')

        # Create a 2x3 grid for multiple charts
        gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

        # Extract data for plotting
        brands = [offer['Brand'] for offer in brand_offers]
        original_prices = [offer['Original_Price'] for offer in brand_offers]
        offer_prices = [offer['Offer_Price'] for offer in brand_offers]
        discounts = [offer['Discount'] for offer in brand_offers]
        ratings = [offer['Rating'] for offer in brand_offers]
        savings = [offer['Savings'] for offer in brand_offers]

        # 1. Price Comparison Bar Chart
        ax1 = fig.add_subplot(gs[0, 0])
        x = np.arange(len(brands))
        width = 0.35

        bars1 = ax1.bar(x - width/2, original_prices, width, label='Original Price',
                       color='#ff6b6b', alpha=0.8)
        bars2 = ax1.bar(x + width/2, offer_prices, width, label='Offer Price',
                       color='#4ecdc4', alpha=0.8)

        ax1.set_title('Price Comparison', fontsize=14, fontweight='bold', color='white')
        ax1.set_xlabel('Brands', color='white')
        ax1.set_ylabel('Price (‚Çπ)', color='white')
        ax1.set_xticks(x)
        ax1.set_xticklabels(brands, rotation=45, ha='right')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Add value labels on bars
        for bar in bars1:
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'‚Çπ{height:.0f}', ha='center', va='bottom', color='white', fontsize=10)

        # 2. Discount Percentage Pie Chart
        ax2 = fig.add_subplot(gs[0, 1])
        colors = plt.cm.Set3(np.linspace(0, 1, len(brands)))
        wedges, texts, autotexts = ax2.pie(discounts, labels=brands, autopct='%1.1f%%',
                                          colors=colors, startangle=90)
        ax2.set_title('Discount Distribution', fontsize=14, fontweight='bold', color='white')

        # 3. Savings vs Rating Scatter Plot
        ax3 = fig.add_subplot(gs[0, 2])
        scatter = ax3.scatter(ratings, savings, c=discounts, s=150, alpha=0.8,
                             cmap='viridis', edgecolors='white', linewidth=2)
        ax3.set_title('Rating vs Savings (Color = Discount%)', fontsize=14, fontweight='bold', color='white')
        ax3.set_xlabel('Rating', color='white')
        ax3.set_ylabel('Savings (‚Çπ)', color='white')
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax3, label='Discount %')

        # Add brand labels to scatter points
        for i, brand in enumerate(brands):
            ax3.annotate(brand, (ratings[i], savings[i]), xytext=(5, 5),
                        textcoords='offset points', fontsize=9, color='white')

        # 4. Offer Value Index (Bottom Left)
        ax4 = fig.add_subplot(gs[1, 0])
        # Calculate value index (higher savings + higher rating = better value)
        value_index = [(savings[i] / max(savings)) * 0.7 + (ratings[i] / 5) * 0.3
                       for i in range(len(brands))]

        bars = ax4.barh(brands, value_index, color='#ff9f43', alpha=0.8)
        ax4.set_title('Smart Value Index', fontsize=14, fontweight='bold', color='white')
        ax4.set_xlabel('Value Score', color='white')
        ax4.grid(True, alpha=0.3)

        # 5. Best Deal Recommendation (Bottom Center)
        ax5 = fig.add_subplot(gs[1, 1])
        ax5.axis('off')

        # Find best deals
        best_discount_idx = discounts.index(max(discounts))
        best_rating_idx = ratings.index(max(ratings))
        best_value_idx = value_index.index(max(value_index))

        recommendation_text = f"""
        üèÜ SMART RECOMMENDATIONS

        üí∞ Best Discount: {brands[best_discount_idx]}
           {discounts[best_discount_idx]}% OFF - Save ‚Çπ{savings[best_discount_idx]}

        ‚≠ê Highest Rated: {brands[best_rating_idx]}
           {ratings[best_rating_idx]}/5 Stars - Premium Quality

        üéØ Best Value: {brands[best_value_idx]}
           Perfect balance of savings & quality
        """

        ax5.text(0.1, 0.5, recommendation_text, fontsize=12, color='#4ecdc4',
                bbox=dict(boxstyle="round,pad=0.3", facecolor='#2c3e50', alpha=0.8),
                transform=ax5.transAxes, verticalalignment='center')

        # 6. Market Share Visualization (Bottom Right)
        ax6 = fig.add_subplot(gs[1, 2])
        units_sold = [offer['Units_Sold'] for offer in brand_offers]
        explode = [0.1 if i == best_value_idx else 0 for i in range(len(brands))]

        ax6.pie(units_sold, labels=brands, autopct='%1.1f%%', explode=explode,
               colors=colors, startangle=90, shadow=True)
        ax6.set_title('Market Share by Sales', fontsize=14, fontweight='bold', color='white')

        plt.tight_layout()
        plt.show()

    def _generate_smart_recommendations(self, product_name, brand_offers):
        """Generate intelligent recommendations based on various factors"""
        print(f"\nü§ñ AI-POWERED RECOMMENDATIONS FOR {product_name.upper()}")
        print("="*70)

        # Sort for different recommendation types
        by_discount = sorted(brand_offers, key=lambda x: x['Discount'], reverse=True)
        by_rating = sorted(brand_offers, key=lambda x: x['Rating'], reverse=True)
        by_value = sorted(brand_offers, key=lambda x: (x['Savings'] + x['Rating']*20), reverse=True)
        by_popularity = sorted(brand_offers, key=lambda x: x['Units_Sold'], reverse=True)

        print("üî• TOP DEALS:")
        print(f"1. MAXIMUM SAVINGS: {by_discount[0]['Brand']} - {by_discount[0]['Discount']}% OFF (Save ‚Çπ{by_discount[0]['Savings']})")
        print(f"2. PREMIUM CHOICE: {by_rating[0]['Brand']} - {by_rating[0]['Rating']}/5 Rating")
        print(f"3. BEST VALUE: {by_value[0]['Brand']} - Great balance of price & quality")
        print(f"4. MOST POPULAR: {by_popularity[0]['Brand']} - {by_popularity[0]['Units_Sold']} units sold")

        # Budget-based recommendations
        print(f"\nüí∞ BUDGET RECOMMENDATIONS:")
        budget_options = sorted(brand_offers, key=lambda x: x['Offer_Price'])
        print(f"‚Ä¢ Budget-Friendly: {budget_options[0]['Brand']} - ‚Çπ{budget_options[0]['Offer_Price']}")
        print(f"‚Ä¢ Mid-Range: {budget_options[len(budget_options)//2]['Brand']} - ‚Çπ{budget_options[len(budget_options)//2]['Offer_Price']}")
        print(f"‚Ä¢ Premium: {budget_options[-1]['Brand']} - ‚Çπ{budget_options[-1]['Offer_Price']}")

        # Stock-based urgency
        low_stock = [offer for offer in brand_offers if offer['Stock'] < 50]
        if low_stock:
            print(f"\n‚ö° LIMITED STOCK ALERT:")
            for item in low_stock[:3]:
                print(f"‚Ä¢ {item['Brand']}: Only {item['Stock']} left! {item['Discount']}% OFF")

    def _add_to_consideration(self, category, product_id, brand_offers):
        """Add product to customer consideration for bill analysis"""
        print(f"\nüõí ADD TO CART SIMULATION:")
        print("Which brand would you like to add to your cart?")
        for i, offer in enumerate(brand_offers[:5], 1):
            print(f"{i}. {offer['Brand']} - ‚Çπ{offer['Offer_Price']} ({offer['Discount']}% OFF)")

        choice = input("Enter choice (1-5) or 0 to skip: ").strip()
        if choice in ['1', '2', '3', '4', '5']:
            selected_offer = brand_offers[int(choice)-1]

            # Add to customer bill simulation
            cart_item = {
                'category': category,
                'product': self.product_data[category][product_id]['name'],
                'brand': selected_offer['Brand'],
                'original_price': selected_offer['Original_Price'],
                'offer_price': selected_offer['Offer_Price'],
                'discount': selected_offer['Discount'],
                'savings': selected_offer['Savings'],
                'rating': selected_offer['Rating'],
                'timestamp': datetime.now()
            }

            # Store in customer bills for analysis
            if not hasattr(self, 'current_cart'):
                self.current_cart = []
            self.current_cart.append(cart_item)

            print(f"‚úÖ Added {selected_offer['Brand']} {self.product_data[category][product_id]['name']} to cart!")
            print(f"   Price: ‚Çπ{selected_offer['Offer_Price']} (‚Çπ{selected_offer['Savings']} savings)")

    def analyze_customer_bills(self):
        """Analyze customer purchase patterns and provide insights"""
        if not hasattr(self, 'current_cart') or not self.current_cart:
            print("\nüõí Your cart is empty! Add some products first.")
            return

        print("\nüìä SMART BILL ANALYSIS & RECOMMENDATIONS")
        print("="*70)

        # Calculate bill summary
        total_original = sum(item['original_price'] for item in self.current_cart)
        total_offer = sum(item['offer_price'] for item in self.current_cart)
        total_savings = sum(item['savings'] for item in self.current_cart)
        avg_rating = sum(item['rating'] for item in self.current_cart) / len(self.current_cart)

        print("üßæ BILL SUMMARY:")
        print("-" * 40)
        for item in self.current_cart:
            print(f"{item['product']} ({item['brand']})")
            print(f"   ‚Çπ{item['original_price']} ‚Üí ‚Çπ{item['offer_price']} ({item['discount']}% OFF)")

        print("-" * 40)
        print(f"Original Total: ‚Çπ{total_original:,.0f}")
        print(f"After Offers:   ‚Çπ{total_offer:,.0f}")
        print(f"Total Savings:  ‚Çπ{total_savings:,.0f} ({((total_savings/total_original)*100):.1f}%)")
        print(f"Avg Rating:     {avg_rating:.1f}/5")

        # Create bill analysis visualization
        self._create_bill_analysis_charts()

        # Generate cross-selling recommendations
        self._generate_cross_sell_recommendations()

    def _create_bill_analysis_charts(self):
        """Create comprehensive bill analysis charts"""
        if not self.current_cart:
            return

        plt.style.use('seaborn-v0_8-darkgrid')
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))
        fig.suptitle('Smart Bill Analysis Dashboard', fontsize=16, fontweight='bold')

        # Extract data
        products = [item['product'] for item in self.current_cart]
        brands = [item['brand'] for item in self.current_cart]
        original_prices = [item['original_price'] for item in self.current_cart]
        offer_prices = [item['offer_price'] for item in self.current_cart]
        savings = [item['savings'] for item in self.current_cart]
        ratings = [item['rating'] for item in self.current_cart]

        # 1. Price comparison
        x = np.arange(len(products))
        width = 0.35

        bars1 = ax1.bar(x - width/2, original_prices, width, label='Original', color='#ff7f0e', alpha=0.8)
        bars2 = ax1.bar(x + width/2, offer_prices, width, label='Offer Price', color='#2ca02c', alpha=0.8)

        ax1.set_title('Price Comparison', fontweight='bold')
        ax1.set_xlabel('Products')
        ax1.set_ylabel('Price (‚Çπ)')
        ax1.set_xticks(x)
        ax1.set_xticklabels([f"{p[:10]}..." if len(p) > 10 else p for p in products], rotation=45)
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # 2. Savings pie chart
        colors = plt.cm.Set3(np.linspace(0, 1, len(products)))
        ax2.pie(savings, labels=brands, autopct='‚Çπ%1.0f', colors=colors, startangle=90)
        ax2.set_title('Savings Breakdown by Brand', fontweight='bold')

        # 3. Rating vs Savings scatter
        ax3.scatter(ratings, savings, s=100, alpha=0.7, c=original_prices, cmap='viridis')
        ax3.set_title('Rating vs Savings', fontweight='bold')
        ax3.set_xlabel('Rating')
        ax3.set_ylabel('Savings (‚Çπ)')
        ax3.grid(True, alpha=0.3)

        # 4. Category distribution
        categories = [item['category'] for item in self.current_cart]
        category_counts = {}
        category_spending = {}

        for item in self.current_cart:
            cat = item['category']
            category_counts[cat] = category_counts.get(cat, 0) + 1
            category_spending[cat] = category_spending.get(cat, 0) + item['offer_price']

        ax4.bar(category_spending.keys(), category_spending.values(), color='steelblue', alpha=0.8)
        ax4.set_title('Spending by Category', fontweight='bold')
        ax4.set_xlabel('Category')
        ax4.set_ylabel('Amount (‚Çπ)')
        ax4.tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.show()

    def _generate_cross_sell_recommendations(self):
        """Generate smart cross-selling recommendations based on cart"""
        if not self.current_cart:
            return

        print("\nüéÅ PERSONALIZED RECOMMENDATIONS BASED ON YOUR CART:")
        print("="*60)

        # Analyze cart categories
        cart_categories = set(item['category'] for item in self.current_cart)

        # Cross-category recommendations
        recommendations = {
            'Grocery': ['Home & Kitchen', 'Bath & Body'],
            'Electronics': ['Home Appliances', 'Footwear'],
            'Clothing': ['Footwear', 'Cosmetics'],
            'Home Appliances': ['Home & Kitchen', 'Electronics'],
            'Cosmetics': ['Bath & Body', 'Clothing'],
            'Bath & Body': ['Cosmetics', 'Home & Kitchen'],
            'Sports & Toys': ['Footwear', 'Clothing'],
            'Footwear': ['Clothing', 'Sports & Toys'],
            'Home & Kitchen': ['Grocery', 'Home Appliances']
        }

        suggested_categories = set()
        for category in cart_categories:
            if category in recommendations:
                suggested_categories.update(recommendations[category])

        # Remove categories already in cart
        suggested_categories -= cart_categories

        # Show top recommendations
        for i, category in enumerate(list(suggested_categories)[:3], 1):
            # Get best offer from suggested category
            best_product_id = list(self.products[category].keys())[0]
            best_offer = self._calculate_best_offer(category, best_product_id)
            product_name = self.products[category][best_product_id]['name']

            print(f"{i}. {category} - Try {product_name} with {best_offer}% OFF!")

        # Seasonal and trending recommendations
        print(f"\nTrending Now:")
        trending_products = [
            ("Electronics", "07", "Smartphone"),
            ("Clothing", "01", "T-Shirt"),
            ("Grocery", "28", "Tea")
        ]

        for cat, prod_id, name in trending_products:
            if cat not in cart_categories:
                offer = self._calculate_best_offer(cat, prod_id)
                print(f"- {name} in {cat} - Up to {offer}% OFF")

    def show_todays_offers(self):
        """Display special daily offers across all categories"""
        print("\nüéÅ TODAY'S SPECIAL OFFERS - LIMITED TIME!")
        print("="*60)

        # Generate special offers
        special_offers = []
        for category, products in self.products.items():
            # Pick 2 random products from each category for special offers
            selected_products = random.sample(list(products.items()), min(2, len(products)))

            for product_id, product_info in selected_products:
                brand_data = self.product_data[category][product_id]['brands']
                best_brand = max(brand_data.items(), key=lambda x: x[1]['discount_potential'])

                special_offers.append({
                    'category': category,
                    'product': product_info['name'],
                    'brand': best_brand[0],
                    'discount': best_brand[1]['discount_potential'] + random.randint(5, 15),  # Extra discount
                    'original_price': best_brand[1]['price'],
                    'rating': best_brand[1]['rating']
                })

        # Sort by discount
        special_offers.sort(key=lambda x: x['discount'], reverse=True)

        # Display top 10 offers
        print("TOP 10 DEALS OF THE DAY:")
        print("-" * 60)
        for i, offer in enumerate(special_offers[:10], 1):
            offer_price = offer['original_price'] * (1 - offer['discount']/100)
            savings = offer['original_price'] - offer_price

            print(f"{i:2d}. {offer['product']} ({offer['brand']})")
            print(f"    {offer['category']} | {offer['discount']}% OFF | "
                  f"‚Çπ{offer['original_price']} ‚Üí ‚Çπ{offer_price:.0f} | Save ‚Çπ{savings:.0f}")
            print(f"    Rating: {offer['rating']}/5")
            print()

    def smart_search(self):
        """Advanced search with filters and recommendations"""
        print("\nüîç SMART PRODUCT SEARCH")
        print("="*50)

        search_term = input("Enter product or brand name: ").strip().lower()
        if not search_term:
            return

        results = []
        for category, products in self.products.items():
            for product_id, product_info in products.items():
                # Search in product names
                if search_term in product_info['name'].lower():
                    best_offer = self._calculate_best_offer(category, product_id)
                    results.append({
                        'type': 'product',
                        'category': category,
                        'product_id': product_id,
                        'name': product_info['name'],
                        'best_offer': best_offer,
                        'brands': len(product_info['brands'])
                    })

                # Search in brand names
                for brand in product_info['brands']:
                    if search_term in brand.lower():
                        brand_data = self.product_data[category][product_id]['brands'][brand]
                        results.append({
                            'type': 'brand',
                            'category': category,
                            'product_id': product_id,
                            'product': product_info['name'],
                            'brand': brand,
                            'discount': brand_data['discount_potential'],
                            'price': brand_data['price'],
                            'rating': brand_data['rating']
                        })

        if not results:
            print(f"No results found for '{search_term}'")
            return

        print(f"\nSearch Results for '{search_term}':")
        print("-" * 50)

        # Display results
        for i, result in enumerate(results[:10], 1):
            if result['type'] == 'product':
                print(f"{i}. {result['name']} ({result['category']})")
                print(f"   {result['brands']} brands available | Best offer: {result['best_offer']}% OFF")
            else:
                offer_price = result['price'] * (1 - result['discount']/100)
                print(f"{i}. {result['brand']} {result['product']} ({result['category']})")
                print(f"   {result['discount']}% OFF | ‚Çπ{result['price']} ‚Üí ‚Çπ{offer_price:.0f} | {result['rating']}/5")

        # Allow selection
        if results:
            choice = input(f"\nSelect item (1-{min(len(results), 10)}) or 0 to go back: ").strip()
            if choice.isdigit() and 1 <= int(choice) <= min(len(results), 10):
                selected = results[int(choice)-1]
                if selected['type'] == 'product':
                    self.display_product_offers(selected['category'], selected['product_id'])

    def generate_analytics_dashboard(self):
        """Comprehensive analytics dashboard"""
        print("\nüìä COMPLETE E-COMMERCE ANALYTICS DASHBOARD")
        print("="*70)

        # Calculate overall metrics
        total_products = sum(len(products) for products in self.products.values())
        total_brands = sum(len(prod_info['brands']) for category in self.products.values()
                          for prod_info in category.values())

        # Category analysis
        category_analysis = {}
        total_revenue = 0

        for category, products in self.products.items():
            cat_revenue = 0
            cat_avg_discount = 0
            cat_products = len(products)
            cat_brands = sum(len(prod_info['brands']) for prod_info in products.values())

            discounts = []
            for product_id in products.keys():
                for brand_data in self.product_data[category][product_id]['brands'].values():
                    cat_revenue += brand_data['revenue']
                    discounts.append(brand_data['discount_potential'])

            cat_avg_discount = sum(discounts) / len(discounts) if discounts else 0

            category_analysis[category] = {
                'revenue': cat_revenue,
                'products': cat_products,
                'brands': cat_brands,
                'avg_discount': cat_avg_discount
            }
            total_revenue += cat_revenue

        # Display category performance
        print("CATEGORY PERFORMANCE ANALYSIS:")
        print("-" * 70)
        print(f"{'Category':<20} {'Revenue':<15} {'Products':<10} {'Brands':<8} {'Avg Discount':<12}")
        print("-" * 70)

        for category, data in sorted(category_analysis.items(), key=lambda x: x[1]['revenue'], reverse=True):
            share = (data['revenue'] / total_revenue) * 100
            print(f"{category:<20} ‚Çπ{data['revenue']:<14,.0f} {data['products']:<10} {data['brands']:<8} {data['avg_discount']:.1f}%")

        print("-" * 70)
        print(f"{'TOTAL':<20} ‚Çπ{total_revenue:<14,.0f} {total_products:<10} {total_brands:<8}")

        # Create comprehensive dashboard
        self._create_analytics_charts(category_analysis, total_revenue)

        # Market insights
        self._display_market_insights(category_analysis)

    def _create_analytics_charts(self, category_analysis, total_revenue):
        """Create comprehensive analytics visualizations"""
        plt.style.use('default')
        fig = plt.figure(figsize=(20, 15))
        fig.suptitle('E-Commerce Analytics Dashboard', fontsize=20, fontweight='bold')

        # Create complex grid layout
        gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

        categories = list(category_analysis.keys())
        revenues = [data['revenue'] for data in category_analysis.values()]
        products = [data['products'] for data in category_analysis.values()]
        brands = [data['brands'] for data in category_analysis.values()]
        discounts = [data['avg_discount'] for data in category_analysis.values()]

        # 1. Revenue by Category (Large chart)
        ax1 = fig.add_subplot(gs[0, :2])
        bars = ax1.bar(categories, revenues, color='steelblue', alpha=0.8)
        ax1.set_title('Revenue by Category', fontsize=16, fontweight='bold')
        ax1.set_ylabel('Revenue (‚Çπ)', fontsize=12)
        ax1.tick_params(axis='x', rotation=45)

        # Add value labels
        for bar, revenue in zip(bars, revenues):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'‚Çπ{revenue/1000000:.1f}M', ha='center', va='bottom')

        # 2. Market Share Pie Chart
        ax2 = fig.add_subplot(gs[0, 2:])
        colors = plt.cm.Set3(np.linspace(0, 1, len(categories)))
        ax2.pie(revenues, labels=categories, autopct='%1.1f%%', colors=colors, startangle=90)
        ax2.set_title('Market Share Distribution', fontsize=16, fontweight='bold')

        # 3. Products vs Brands Correlation
        ax3 = fig.add_subplot(gs[1, 0])
        ax3.scatter(products, brands, s=[r/100000 for r in revenues], alpha=0.7, c=discounts, cmap='viridis')
        ax3.set_title('Products vs Brands', fontweight='bold')
        ax3.set_xlabel('Products')
        ax3.set_ylabel('Brands')

        # 4. Average Discount by Category
        ax4 = fig.add_subplot(gs[1, 1])
        bars = ax4.barh(categories, discounts, color='orange', alpha=0.8)
        ax4.set_title('Avg Discount %', fontweight='bold')
        ax4.set_xlabel('Discount %')

        # 5. Revenue Trend (Simulated)
        ax5 = fig.add_subplot(gs[1, 2:])
        months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug']
        trend = [total_revenue * random.uniform(0.7, 1.3) for _ in months]
        ax5.plot(months, trend, marker='o', linewidth=3, markersize=8, color='red')
        ax5.fill_between(months, trend, alpha=0.3, color='red')
        ax5.set_title('Revenue Trend (2024)', fontweight='bold')
        ax5.set_ylabel('Revenue (‚Çπ)')
        ax5.grid(True, alpha=0.3)

        # 6. Top Performing Products
        ax6 = fig.add_subplot(gs[2, :2])
        top_products = []
        for category, products in self.products.items():
            for product_id, product_info in products.items():
                total_prod_revenue = sum(brand['revenue'] for brand in
                                       self.product_data[category][product_id]['brands'].values())
                top_products.append((product_info['name'], total_prod_revenue))

        top_products.sort(key=lambda x: x[1], reverse=True)
        top_names = [p[0] for p in top_products[:8]]
        top_revenues = [p[1] for p in top_products[:8]]

        bars = ax6.barh(top_names, top_revenues, color='green', alpha=0.8)
        ax6.set_title('Top 8 Products by Revenue', fontweight='bold')
        ax6.set_xlabel('Revenue (‚Çπ)')

        # 7. Offer Distribution
        ax7 = fig.add_subplot(gs[2, 2:])
        all_discounts = []
        for category in self.product_data.values():
            for product in category.values():
                for brand_data in product['brands'].values():
                    all_discounts.append(brand_data['discount_potential'])

        ax7.hist(all_discounts, bins=20, color='purple', alpha=0.7, edgecolor='black')
        ax7.set_title('Discount Distribution', fontweight='bold')
        ax7.set_xlabel('Discount %')
        ax7.set_ylabel('Frequency')
        ax7.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def _display_market_insights(self, category_analysis):
        """Display strategic market insights"""
        print(f"\nüí° STRATEGIC MARKET INSIGHTS")
        print("="*50)

        # Find best and worst performing categories
        sorted_categories = sorted(category_analysis.items(), key=lambda x: x[1]['revenue'], reverse=True)
        best_category = sorted_categories[0]
        worst_category = sorted_categories[-1]

        print(f"üèÜ TOP PERFORMER: {best_category[0]}")
        print(f"   Revenue: ‚Çπ{best_category[1]['revenue']:,.0f}")
        print(f"   Market Share: {(best_category[1]['revenue']/sum(data['revenue'] for data in category_analysis.values()))*100:.1f}%")

        print(f"\nüìà GROWTH OPPORTUNITY: {worst_category[0]}")
        print(f"   Current Revenue: ‚Çπ{worst_category[1]['revenue']:,.0f}")
        print(f"   Avg Discount: {worst_category[1]['avg_discount']:.1f}%")
        print(f"   Recommendation: Increase promotional activities")

        # Discount analysis
        high_discount_categories = [cat for cat, data in category_analysis.items() if data['avg_discount'] > 25]
        if high_discount_categories:
            print(f"\nüî• HIGH DISCOUNT CATEGORIES: {', '.join(high_discount_categories)}")
            print("   Strategy: Leverage for customer acquisition")

    def run(self):
        """Main program execution"""
        print("Welcome to the Smart E-Commerce Offer Recommendation System!")
        print("Discover personalized offers and smart shopping insights.")

        while True:
            self.display_main_menu()

            try:
                choice = input("\nEnter your choice (0-9): ").strip()

                if choice == '0':
                    print("Thank you for using Smart Offer Recommendation System!")
                    break
                elif choice == '9':
                    self.generate_analytics_dashboard()
                elif choice == '8':
                    self.analyze_customer_bills()
                elif choice == '7':
                    self.show_todays_offers()
                elif choice == '6':
                    self.smart_search()
                elif choice in ['1', '2', '3', '4', '5']:
                    category_index = int(choice) - 1
                    categories = list(self.products.keys())
                    if category_index < len(categories):
                        selected_category = categories[category_index]

                        while True:
                            self.display_category_products(selected_category)
                            product_choice = input(f"\nSelect a product from {selected_category} (0 to go back): ").strip()

                            if product_choice == '0':
                                break
                            elif product_choice.zfill(2) in self.products[selected_category]:
                                self.display_product_offers(selected_category, product_choice.zfill(2))
                                input("\nPress Enter to continue...")
                            else:
                                print("Invalid product selection!")
                else:
                    print("Invalid choice! Please select 0-9.")

                if choice != '0':
                    input("\nPress Enter to continue...")

            except (ValueError, KeyboardInterrupt):
                print("\nInvalid input! Please try again.")

# Run the system
if __name__ == "__main__":
    system = SmartOfferRecommendationSystem()
    system.run()
```
